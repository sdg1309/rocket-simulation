```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
datos = pd.read_csv('2D_Simulation.csv', header=0)
df = datos.to_numpy()
datos.head()
```

```python
xcol = df[:,2]
ycol = df[:,3]
plt.plot(xcol, ycol)
plt.grid(True)
plt.xlabel('x')
plt.xticks(rotation = 25)
plt.ylabel('y (in hundreds of millions)')
plt.yticks(rotation = 25)
plt.title("Trajectory")
plt.show()
```

```python
xcol = df[:,0]
ycol = df[:,8]
plt.plot(xcol, ycol)
plt.grid(True)
plt.xlabel('time')
plt.xticks(rotation = 25)
plt.ylabel('Acceleration in y')
plt.yticks(rotation = 25)
plt.title("General Acceleration in the Y-Axis")
plt.show()
```
